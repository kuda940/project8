```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')
df['Day of Week'] = df['Date'].dt.day_name()
display(df.head())
```

```python
df['Months_name'] = df['Date'].dt.month_name()
display(df.head())
```

```python
df['week number']=df['Date'].dt.isocalendar().week
display(df.head())
```

```python
# create bins on price category

bins = [0, 50, 200, 500, 1000, np.inf]
labels = ['0-50', '51-200', '201-500', '501-1000', '1000+']
df['Price_Category'] = pd.cut(df['Price'], bins=bins, labels=labels, right=True)

display(df.head())
```

```python
# Filter the DataFrame for 'Completed' orders
filtered_df = df[df['Status'] == 'Completed']

# Display the first 5 rows of the filtered DataFrame
display(filtered_df.head())

print(f"Number of 'Completed' orders: {len(filtered_df)}")
```

```python
unique_statuses = df['Status'].unique()
print(f"Unique statuses in the 'Status' column: {unique_statuses}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.countplot(data=df, x='Category', hue='Price_Category', palette='viridis',
              order=df['Category'].value_counts().index)
plt.title('Relationship between Product Category and Price Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Price Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# create bins using quantity

bins_qty = [0, 1, 2, 3, 4, 5, np.inf]
labels_qty = ['1', '2', '3', '4', '5', '6+']
df['Quantity_Category'] = pd.cut(df['Quantity'], bins=bins_qty, labels=labels_qty, right=True)

display(df.head())
```

```python
# create bins for total sales

bins_sales = [0, 100, 500, 1000, 2000, np.inf]
labels_sales = ['0-100', '101-500', '501-1000', '1001-2000', '2000+']
df['Sales_Category'] = pd.cut(df['Total Sales'], bins=bins_sales, labels=labels_sales, right=True)

display(df.head())
```

## Filter Data for Top Products

### Subtask:
Filter the main DataFrame `df` to include only the sales data for the top 5 products identified previously.


**Reasoning**:
To filter the DataFrame, I first need to extract the names of the top 5 products from the `top_5_products` Series index and convert them into a list. This list will then be used to filter the main `df` DataFrame.



```python
top_product_names = top_5_products.index.tolist()
top_products_df = df[df['Product'].isin(top_product_names)]

display(top_products_df.head())
print(f"Number of rows in top_products_df: {len(top_products_df)}")
print(f"Unique products in top_products_df: {top_products_df['Product'].unique().tolist()}")
```

**Reasoning**:
Now that the data has been filtered for the top 5 products, the next step is to group this data by 'Month_Number' and 'Product' and calculate the sum of 'Total Sales' to get the monthly sales for each top product. This will prepare the data for the requested visualization.



```python
monthly_sales_top_products = top_products_df.groupby(['Month_Number', 'Product'])['Total Sales'].sum().reset_index()
display(monthly_sales_top_products.head())
print(f"Shape of monthly_sales_top_products: {monthly_sales_top_products.shape}")
```

**Reasoning**:
To visualize the monthly sales trends for each of the top 5 products, I will create a line plot using `seaborn.lineplot`. Each product will be represented by a different line, making it easy to compare their sales performance over the months.



```python
plt.figure(figsize=(12, 7))
sns.lineplot(data=monthly_sales_top_products, x='Month_Number', y='Total Sales', hue='Product', marker='o')
plt.title('Monthly Total Sales for Top 5 Products')
plt.xlabel('Month Number')
plt.ylabel('Total Sales')
plt.xticks(monthly_sales_top_products['Month_Number'].unique())
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Summary of Sales Trends for Top 5 Products

Based on the line plot showing the monthly total sales for the top 5 products, the following trends can be observed:

*   **Refrigerator**: Shows a strong performance in February, with sales slightly decreasing in March. It maintains the highest total sales overall.
*   **Laptop**: Sales are relatively high in February, with a slight dip in March.
*   **Smartphone**: Similar to Refrigerator and Laptop, Smartphone sales are higher in February and show a decrease in March.
*   **Washing Machine**: Sales for Washing Machine are consistent across both months, but significantly lower compared to the top three products.
*   **Smartwatch**: Smartwatch sales are the lowest among the top 5 products and remain fairly stable or slightly increase from February to March.

**Overall Trends:**
*   There appears to be a general **decrease in sales from February to March** for the top-performing products like Refrigerator, Laptop, and Smartphone. This could indicate a seasonal pattern, with February potentially being a stronger sales month for these categories, or it could be related to the limited time frame of the data (only two months).
*   The **Refrigerator** is consistently the highest revenue generator, followed by Laptop and Smartphone.
*   **Washing Machine** and **Smartwatch** have lower total sales compared to the other top products, suggesting they contribute less to overall revenue, at least in these observed months. Their sales figures are also more stable.

Further analysis with more months of data would be beneficial to confirm seasonal patterns and longer-term trends.
