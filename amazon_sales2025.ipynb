# Task
Analyze sales by day of the week, month, and day of the month. Summarize and compare the findings across these date-related columns.

## Analyze sales by day of the week

### Subtask:
Calculate and visualize the total sales for each day of the week.


**Reasoning**:
Group the DataFrame by 'Day Name', sum 'Total Sales', sort the days, and create a bar plot to visualize the total sales per day of the week.



```python
sales_by_day = df.groupby('Day Name')['Total Sales'].sum().reset_index()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_by_day['Day Name'] = pd.Categorical(sales_by_day['Day Name'], categories=day_order, ordered=True)
sales_by_day = sales_by_day.sort_values('Day Name')

plt.figure(figsize=(10, 6))
sns.barplot(x='Day Name', y='Total Sales', data=sales_by_day, palette='viridis')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.title('Total Sales by Day of the Week')
plt.show()
```

## Analyze sales by month

### Subtask:
Calculate and visualize the total sales for each month.


**Reasoning**:
Calculate the total sales for each month and visualize the results using a bar plot to understand monthly sales performance.



```python
sales_by_month = df.groupby('month name')['Total Sales'].sum().reset_index()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales_by_month['month name'] = pd.Categorical(sales_by_month['month name'], categories=month_order, ordered=True)
sales_by_month = sales_by_month.sort_values('month name')

plt.figure(figsize=(12, 6))
sns.barplot(x='month name', y='Total Sales', data=sales_by_month, palette='viridis')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Analyze sales by day of the month

### Subtask:
Calculate and visualize the total sales for each day of the month.


**Reasoning**:
Group the DataFrame by 'week end' and calculate the sum of 'Total Sales', then sort by 'week end' and create a bar plot to visualize the total sales by day of the month.



```python
sales_by_day_of_month = df.groupby('week end')['Total Sales'].sum().reset_index()
sales_by_day_of_month = sales_by_day_of_month.sort_values('week end')

plt.figure(figsize=(15, 7))
sns.barplot(x='week end', y='Total Sales', data=sales_by_day_of_month, palette='viridis')
plt.xlabel('Day of the Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Day of the Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
